```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open ("01_input.txt", "r") as myfile:
    data=myfile.read()
data=data.strip()
data=data.split(", ")
data=[[i[0],int(i[1:])] for i in data]

pos = 0 + 0j
dir = 0 + 1j
hist=[pos]
for i in data:
    if i[0] == 'L':
        dir*=1j
    else: dir*=-1j
    for j in range(1,i[1]+1):
        pos+=dir
        hist.append(pos)
pos=pos.real + pos.imag
print(pos)

for i in range(len(hist)):
    if hist[i] in hist[i+1:]:
        print(hist[i])
        break
```

```python
with open ("02_input.txt", "r") as myfile:
    data=myfile.readlines()
data=[item.strip() for item in data]
data

for i in data:
    [x,y]=[0,0];
    for j in range(len(i)):
        if i[j]=='L' and x>-1:
            x-=1
        if i[j]=='R' and x<1:
            x+=1
        if i[j]=='U' and y<1:
            y+=1
        if i[j]=='D' and y>-1:
            y-=1
    print([x,y])
print([1,2,3])
print([4,5,6])
print([7,8,9])
print("\n")

for i in data:
    [x,y]=[-2,0];
    for j in range(len(i)):
        if i[j]=='L' and abs(x-1)+abs(y)<=2:
            x-=1
        if i[j]=='R' and abs(x+1)+abs(y)<=2:
            x+=1
        if i[j]=='U' and abs(x)+abs(y+1)<=2:
            y+=1
        if i[j]=='D' and abs(x)+abs(y-1)<=2:
            y-=1
    print([x,y])
        
        
PAD = [['x', 'x', '1', 'x', 'x'],
       ['x', '2', '3', '4', 'x'],
       ['5', '6', '7', '8', '9'],
       ['x', 'A', 'B', 'C', 'x'],
       ['x', 'x', 'D', 'x', 'x'],
       ]
print(PAD)
```

```python
data = np.loadtxt("3_input.txt")
p=0
for i in data:
    i=np.sort(i)
    if i[0]+i[1]>i[2]:
        p+=1
print(p)

data_t=np.transpose(data)

p2=0
for i in range(0,3):
    j=0
    while j<1599:
        dat=[data_t[i,j],data_t[i,j+1],data_t[i,j+2]]
        dat=np.sort(dat)
        if dat[0]+dat[1]>dat[2]:
            p2+=1
        j+=3
print(p2)
```
